## Numpy for Images

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

color_image = plt.imread("images/tree.jpeg")
color_image_copy = color_image.copy()
img = color_image.copy()

print(f'Color Image Shape: {color_image_copy.shape}')
print(f'Color Image Size (Shape multiplication): {color_image_copy.size}')
print(f'Color Image dimesions:{color_image_copy.ndim}')
print(f'Color Image 1st Pixel value = {color_image_copy[1][0]}')

color_image_copy[:, 340:370] = [255, 0, 255]
print(f'Color Image 1st Pixel after updating the value = {color_image_copy[0][0]}')
plt.imshow(color_image_copy)
plt.show()
```

```python
img[:, 250:270, 0] = 255
img[:, 300:320, 1] = 255
img[:, 350:370, 2] = 255
plt.imshow(img)
```

```python
gray = lambda img : np.dot(img[... , :3] , [0.299 , 0.587, 0.114]) 
gray_image = gray(color_image)
```

```python
plt.imshow(gray_image, cmap = plt.get_cmap(name = 'gray'))
print(gray_image)
print(f'Gray Image dimensions = {gray_image.ndim}')
print(f'Gray Image shape = {gray_image.shape}')
print(f'Length of Gray Image columns = {len(gray_image[0])}')
print(f'Data Type of Gray Image = {gray_image.dtype}')
```

```python
mask_test_image = color_image.copy()
print(f'Existing Flags = {mask_test_image.flags}')
rows, columns, _ = mask_test_image.shape
print(f'Rows = {rows}, Columns = {columns}')
print(mask_test_image[0][0])
 
mask_test_image[0:50,0:50] = 128
mask_test_image[rows-50:rows,columns-50:columns] = 128
mask_test_image[rows-50:rows,0:50] = 128
mask_test_image[rows-50:rows,0:50] = 128
mask_test_image[0:50, columns-50:columns] = [128, 128, 128]
plt.imshow(mask_test_image)
```

```python
vertical_gradient = np.linspace(0, 255, 2500, dtype=np.uint8).reshape((50, 50))
print(f'Vertical Gradient Shape = {vertical_gradient.shape}')
print(f'Vertical Gradient size = {vertical_gradient.size}')
print(f'Vertical Gradient dimension = {vertical_gradient.ndim}')
print(vertical_gradient)

plt.imshow(vertical_gradient, cmap='gray')
plt.show()
```

```python
vertical_right_gradient = np.flip(vertical_gradient, axis=-1)
print(vertical_right_gradient)
plt.imshow(vertical_right_gradient, cmap='gray')
```

```python
horizontal_left_gradient = np.rot90(vertical_gradient)
print(horizontal_left_gradient)
plt.imshow(horizontal_left_gradient, cmap='gray')
```

```python
horizontal_left_gradient = np.rot90(vertical_gradient)
print(horizontal_left_gradient)
horizontal_right_gradient = np.flipud(horizontal_left_gradient)
print(horizontal_right_gradient)
plt.imshow(horizontal_right_gradient, cmap='gray')
```

## Histograms

```python
import matplotlib.pyplot as plt
import numpy as np


image = plt.imread("images/tree.jpeg")
red_plane = image[:,:,0]
green_plane = image[:,:,1]
blue_plane = image[:,:,2]

plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(image)

hist, bins = np.histogram(red_plane.ravel(), bins=256, range=(0,256))
plt.subplot(2, 2, 2)
plt.title("Red Histogram")
plt.bar(bins[:-1], hist)

hist, bins = np.histogram(green_plane.ravel(), bins=256, range=(0,256))
plt.subplot(2, 2, 3)
plt.title("Green Histogram")
plt.bar(bins[:-1], hist)

hist, bins = np.histogram(blue_plane.ravel(), bins=256, range=(0,256))
plt.subplot(2, 2, 4)
plt.title("Blue Histogram")
plt.bar(bins[:-1], hist)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.title("Original Image using Matplotlib")
plt.imshow(image)

plt.subplots_adjust(hspace=1, wspace=1)

plt.subplot(2,2,2)
plt.title("Green Histogram using Matplotlib")
plt.hist(green_plane.ravel(), bins=256, range=(0, 256))

plt.subplot(2,2,3)
plt.title("Red Histogram using Matplotlib")
plt.hist(green_plane.ravel(), bins=256, range=(0, 256))

plt.subplot(2,2,4)
plt.title("Blue Histogram using Matplotlib")
plt.hist(blue_plane.ravel(), bins=256, range=(0, 256))

plt.show()
```
