```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(5)
print(x)
```

```python
y = x
print(y)
```

```python
plt.plot(x, y, 'o-')
plt.plot(x, -y, 'o--')
plt.title('y=x and y=-x')
plt.show()
```

```python
N = 11
x = np.linspace(0, 10, 5)  # N = number of points (auto cal the steps)
y = x
print(y)
```

```python
plt.plot(x, y, 'o--')
plt.axis('off')
plt.show()
```

```python
y = np.logspace(0.1, 1, N)
print(y)
plt.plot(y, y, 'o--')
plt.show()
```

```python
y = np.geomspace(0.1, 1000, N)
print(y)
plt.plot(y, y, 'o--')
plt.show()
```

```python
x = np.arange(10)
plt.plot(x)
plt.show()
```

```python
plt.plot(x, [y**2 for y in x])  # complex way
plt.show()
```

```python
plt.plot(x, x**2)  # easy way
plt.show()
```

## Multiline plots

```python
plt.plot(x, x**2)  # easy way
plt.plot(x, x**3) 
plt.plot(x, x*2) 
plt.show()
```

```python
plt.plot(x, x**2, x, x**3, x,  x*2)  # same thing in single line
plt.show()
```

## Grid Axis Labels

```python
x = np.arange(3)
plt.plot(x, x**2, x, x**3, x, x*2)
plt.grid(True)
print(plt.axis())
plt.show()
```

```python
plt.plot(x, x**2, x, x**3, x, x*2)
plt.grid(True)
plt.axis([0, 2, 0, 10])  # changing the axis start and end points
print(plt.axis())
plt.show()
```

```python
plt.plot(x, x**2, x, x**3, x, x*2)
plt.grid(True)
plt.xlim([0, 2])  # same thing can be achievved  using limits
plt.ylim([0, 12])
print(plt.axis())
plt.show()
```

```python
plt.plot(x, x**2, x, x**3, x, x*2)
plt.ylabel('y axis label')
plt.xlabel('x axis label')
plt.title("this is my title")
 
plt.show()
```

```python
plt.plot(x, x**2, label='first')
plt.plot(x, x**3, label='second')
plt.plot(x, x*2, label='third')
plt.legend()
```

```python
# Another way of legend defining and changing location
plt.plot(x, x**2, x, x**3, x, x*2)
plt.legend(['first', 'second', 'third'], loc='upper center')
plt.savefig('test.png')
```

## IO

```python
import matplotlib.pyplot as plt

img1 = plt.imread("images/tree.jpeg")  # matplotlib only supports png
# for other types we need Pillow library to install

plt.imshow(img1)
plt.axis('off')
plt.title('My Tree')
plt.show()
plt.imsave("tree.jpg", img1)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
img2 = Image.open("images/tree.jpeg").convert("L")
plt.imshow(img2, cmap = "gray")
plt.axis("off")
plt.savefig("tree_gray.png")
plt.show()
```
