```python
import skimage.data as data


print(dir(data))
img  = data.chelsea()
io.imshow(img)
```

```python
import matplotlib.pyplot as plt

img = data.coffee()

plt.imshow(img)
plt.title('Coffee')
plt.axis('off')
plt.show()
```

```python
img1 = data.coffee()
img2 = data.rocket()
img3 = data.horse()
img4 = data.astronaut()

titles = ['Coffee', 'Rocket', 'Horse', 'Astronaut']
images = [img1, img2, img3, img4]

for i in range(4):
    plt.subplot(2, 2, i+1)
    if i == 2:
        plt.imshow(images[i], cmap='gray')
    else:
         plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
