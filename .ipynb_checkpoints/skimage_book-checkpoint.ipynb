## Canny Edge Detection

```python
import numpy as np
from skimage import feature, data, io, color


image = data.rocket()
io.imshow(image)


 
```

```python
gray = color.rgb2gray(image)
io.imshow(gray)
print(type(gray))
```

```python
edges = feature.canny(gray, sigma=5)
io.imshow(edges)
```

```python
page1_gray = color.rgb2gray(io.imread("page1.tif"))
io.imshow(page1_gray)
page1_edge = feature.canny(page1_gray, sigma=3)
print(type(page1_edge))

from skimage.util import img_as_uint

edged_converted = img_as_uint(page1_edge)


io.imsave("check1.jpg", edged_converted)
```
